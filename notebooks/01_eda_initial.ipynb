```python
# notebooks/01_eda_initial.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Configuraci√≥n de estilo
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.float_format', '{:,.2f}'.format)

# Cargar datos
df = pd.read_csv("../data/raw/rent_guatemala.csv")

print("üîç DIMENSIONES Y TIPOS DE DATOS")
print(f"Dataset: {df.shape[0]} filas, {df.shape[1]} columnas")
print("\nüìã Primeras filas:")
display(df.head(3))

print("\nüìä INFORMACI√ìN GENERAL")
print(df.info())
print("\nüìà ESTAD√çSTICAS DESCRIPTIVAS")
display(df.describe())

# 1. AN√ÅLISIS DE LA VARIABLE TARGET
print("üéØ AN√ÅLISIS DE PRECIOS (rent_price_gtq)")
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Distribuci√≥n general
sns.histplot(df['rent_price_gtq'], bins=50, ax=axes[0,0])
axes[0,0].set_title('Distribuci√≥n de Precios de Alquiler')
axes[0,0].axvline(df['rent_price_gtq'].mean(), color='red', linestyle='--', label=f'Media: Q{df["rent_price_gtq"].mean():.0f}')
axes[0,0].legend()

# Boxplot
sns.boxplot(y=df['rent_price_gtq'], ax=axes[0,1])
axes[0,1].set_title('Boxplot - Precios de Alquiler')

# Distribuci√≥n logar√≠tmica (para mejor visualizaci√≥n si hay outliers)
df['log_price'] = np.log1p(df['rent_price_gtq'])
sns.histplot(df['log_price'], bins=50, ax=axes[1,0])
axes[1,0].set_title('Distribuci√≥n Logar√≠tmica de Precios')

# Precio por m2
sns.histplot(df['price_per_m2'], bins=50, ax=axes[1,1])
axes[1,1].set_title('Distribuci√≥n de Precio por m¬≤')

plt.tight_layout()
plt.show()

# Estad√≠sticas de precios
print(f"üí∞ Rango de precios: Q{df['rent_price_gtq'].min():,.0f} - Q{df['rent_price_gtq'].max():,.0f}")
print(f"üìä Precio promedio: Q{df['rent_price_gtq'].mean():,.0f}")
print(f"üìà Precio mediano: Q{df['rent_price_gtq'].median():,.0f}")

# 2. AN√ÅLISIS GEOGR√ÅFICO DETALLADO
print("\nüó∫Ô∏è AN√ÅLISIS GEOGR√ÅFICO POR ZONAS")
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Precio promedio por zona (top 15)
zone_analysis = df.groupby('zone').agg({
    'rent_price_gtq': ['mean', 'count', 'std'],
    'size_m2': 'mean',
    'price_per_m2': 'mean'
}).round(0)
zone_analysis.columns = ['precio_promedio', 'cantidad', 'std_precio', 'tama√±o_promedio', 'precio_m2_promedio']
zone_analysis = zone_analysis.sort_values('precio_promedio', ascending=False)

zone_analysis['precio_promedio'].head(15).plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Precio Promedio por Zona (Top 15)')
axes[0,0].tick_params(axis='x', rotation=45)

# Cantidad de propiedades por zona
zone_analysis['cantidad'].sort_values(ascending=False).head(15).plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Cantidad de Propiedades por Zona (Top 15)')
axes[0,1].tick_params(axis='x', rotation=45)

# Precio por m2 por zona
zone_analysis['precio_m2_promedio'].sort_values(ascending=False).head(15).plot(kind='bar', ax=axes[1,0], color='salmon')
axes[1,0].set_title('Precio por m¬≤ por Zona (Top 15)')
axes[1,0].tick_params(axis='x', rotation=45)

# Relaci√≥n tama√±o vs precio por zona
scatter = axes[1,1].scatter(zone_analysis['tama√±o_promedio'], zone_analysis['precio_promedio'], 
                           s=zone_analysis['cantidad']/10, alpha=0.6)
axes[1,1].set_xlabel('Tama√±o Promedio (m¬≤)')
axes[1,1].set_ylabel('Precio Promedio (GTQ)')
axes[1,1].set_title('Relaci√≥n: Tama√±o vs Precio por Zona\n(Tama√±o del bubble = cantidad de propiedades)')

# Etiquetar algunos puntos interesantes
for zone in zone_analysis.index[:5]:  # Top 5 zonas m√°s caras
    axes[1,1].annotate(zone, 
                      (zone_analysis.loc[zone, 'tama√±o_promedio'], zone_analysis.loc[zone, 'precio_promedio']),
                      xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

display(zone_analysis.head(10))

# 3. AN√ÅLISIS POR TIPO DE PROPIEDAD
print("\nüè† AN√ÅLISIS POR TIPO DE PROPIEDAD")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribuci√≥n de tipos
df['property_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0])
axes[0].set_title('Distribuci√≥n de Tipos de Propiedad')

# Precio por tipo
sns.boxplot(data=df, x='property_type', y='rent_price_gtq', ax=axes[1])
axes[1].set_title('Distribuci√≥n de Precios por Tipo de Propiedad')
axes[1].tick_params(axis='x', rotation=45)

# Tama√±o por tipo
sns.boxplot(data=df, x='property_type', y='size_m2', ax=axes[2])
axes[2].set_title('Distribuci√≥n de Tama√±os por Tipo de Propiedad')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 4. AN√ÅLISIS DE CORRELACIONES
print("\nüìä AN√ÅLISIS DE CORRELACIONES")

# Matriz de correlaci√≥n
plt.figure(figsize=(14, 10))
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={"shrink": .8})
plt.title('Matriz de Correlaci√≥n (Triangular)')
plt.tight_layout()
plt.show()

# Correlaciones con el precio (ordenadas)
price_correlations = correlation_matrix['rent_price_gtq'].sort_values(ascending=False)
print("üîó CORRELACIONES CON EL PRECIO:")
display(pd.DataFrame(price_correlations).head(10))

# 5. AN√ÅLISIS DE CARACTER√çSTICAS PRINCIPALES
print("\nüìè AN√ÅLISIS DE CARACTER√çSTICAS F√çSICAS")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Tama√±o vs Precio
axes[0,0].scatter(df['size_m2'], df['rent_price_gtq'], alpha=0.5, s=1)
axes[0,0].set_xlabel('Tama√±o (m¬≤)')
axes[0,0].set_ylabel('Precio (GTQ)')
axes[0,0].set_title('Relaci√≥n: Tama√±o vs Precio')

# Habitaciones vs Precio
df.boxplot(column='rent_price_gtq', by='bedrooms', ax=axes[0,1])
axes[0,1].set_title('Precio por N√∫mero de Habitaciones')
axes[0,1].set_xlabel('Habitaciones')

# Ba√±os vs Precio
df.boxplot(column='rent_price_gtq', by='bathrooms', ax=axes[0,2])
axes[0,2].set_title('Precio por N√∫mero de Ba√±os')
axes[0,2].set_xlabel('Ba√±os')

# Antig√ºedad vs Precio
axes[1,0].scatter(df['age_years'], df['rent_price_gtq'], alpha=0.5, s=1)
axes[1,0].set_xlabel('Antig√ºedad (a√±os)')
axes[1,0].set_ylabel('Precio (GTQ)')
axes[1,0].set_title('Relaci√≥n: Antig√ºedad vs Precio')

# Piso vs Precio (solo apartamentos)
apt_data = df[df['property_type'] == 'Apartamento']
if not apt_data.empty:
    axes[1,1].scatter(apt_data['floor'], apt_data['rent_price_gtq'], alpha=0.5, s=1)
    axes[1,1].set_xlabel('Piso')
    axes[1,1].set_ylabel('Precio (GTQ)')
    axes[1,1].set_title('Relaci√≥n: Piso vs Precio (Solo Apartamentos)')

# Distribuci√≥n de antig√ºedad
df['age_years'].hist(bins=30, ax=axes[1,2])
axes[1,2].set_xlabel('Antig√ºedad (a√±os)')
axes[1,2].set_ylabel('Frecuencia')
axes[1,2].set_title('Distribuci√≥n de Antig√ºedad')

plt.tight_layout()
plt.show()

# 6. AN√ÅLISIS DE AMENITIES
print("\n‚≠ê AN√ÅLISIS DE AMENITIES Y SERVICIOS")
amenities = ['has_security', 'has_pool', 'has_gym', 'has_social', 'furnished', 
             'balcony', 'garden', 'pet_friendly', 'has_elevator']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, amenity in enumerate(amenities):
    if i < len(axes):
        # Precio promedio con/sin amenity
        amenity_effect = df.groupby(amenity)['rent_price_gtq'].mean()
        amenity_effect.plot(kind='bar', ax=axes[i], color=['lightcoral', 'lightgreen'])
        axes[i].set_title(f'Precio Promedio: {amenity}')
        axes[i].set_xticklabels(['No', 'S√≠'], rotation=0)
        
        # Calcular diferencia porcentual
        if len(amenity_effect) == 2:
            diff_pct = (amenity_effect[1] - amenity_effect[0]) / amenity_effect[0] * 100
            axes[i].text(0.5, 0.9, f'+{diff_pct:.1f}%', transform=axes[i].transAxes, 
                        ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

# 7. AN√ÅLISIS DE ESTACIONAMIENTO
print("\nüöó AN√ÅLISIS DE ESTACIONAMIENTO")
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Precio por espacios de estacionamiento
df.boxplot(column='rent_price_gtq', by='parking_spaces', ax=axes[0])
axes[0].set_title('Precio por Espacios de Estacionamiento')
axes[0].set_xlabel('Espacios de Estacionamiento')

# Distribuci√≥n de espacios
df['parking_spaces'].value_counts().sort_index().plot(kind='bar', ax=axes[1])
axes[1].set_title('Distribuci√≥n de Espacios de Estacionamiento')
axes[1].set_xlabel('Espacios de Estacionamiento')
axes[1].set_ylabel('Cantidad')

plt.tight_layout()
plt.show()

# 8. AN√ÅLISIS MULTIVARIADO
print("\nüîç AN√ÅLISIS MULTIVARIADO")

# Precio vs Tama√±o, coloreado por tipo de propiedad
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['size_m2'], df['rent_price_gtq'], 
                     c=df['bedrooms'], alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='N√∫mero de Habitaciones')
plt.xlabel('Tama√±o (m¬≤)')
plt.ylabel('Precio (GTQ)')
plt.title('Relaci√≥n: Tama√±o vs Precio (Coloreado por Habitaciones)')
plt.show()

# 9. RESUMEN DE INSIGHTS
print("\nüí° PRINCIPALES INSIGHTS DEL EDA")

print("1. üìä DISTRIBUCI√ìN DE PRECIOS:")
print(f"   - Rango: Q{df['rent_price_gtq'].min():,.0f} - Q{df['rent_price_gtq'].max():,.0f}")
print(f"   - Mediana: Q{df['rent_price_gtq'].median():,.0f}")

print("\n2. üó∫Ô∏è PATRONES GEOGR√ÅFICOS:")
print(f"   - Zona m√°s cara: {zone_analysis.index[0]} (Q{zone_analysis['precio_promedio'].iloc[0]:,.0f})")
print(f"   - Zona m√°s econ√≥mica: {zone_analysis.index[-1]} (Q{zone_analysis['precio_promedio'].iloc[-1]:,.0f})")

print("\n3. üè† TIPO DE PROPIEDAD:")
print(f"   - Apartamentos: {len(df[df['property_type']=='Apartamento'])} propiedades")
print(f"   - Casas: {len(df[df['property_type']=='Casa'])} propiedades")

print("\n4. üîó VARIABLES M√ÅS CORRELACIONADAS CON EL PRECIO:")
top_corr = price_correlations.head(6)
for var, corr in top_corr.items():
    if var != 'rent_price_gtq':
        print(f"   - {var}: {corr:.3f}")

print("\n5. ‚≠ê AMENITIES M√ÅS VALORADAS:")
for amenity in amenities:
    if df[amenity].nunique() == 2:
        with_amenity = df[df[amenity] == 1]['rent_price_gtq'].mean()
        without_amenity = df[df[amenity] == 0]['rent_price_gtq'].mean()
        if with_amenity > without_amenity:
            diff_pct = (with_amenity - without_amenity) / without_amenity * 100
            print(f"   - {amenity}: +{diff_pct:.1f}% de premium")

print("\nüéØ CONCLUSIONES PARA EL MODELADO:")
print("   - Variables clave: zona, tama√±o, habitaciones, amenities premium")
print("   - Considerar transformaciones logar√≠tmicas para precios")
print("   - Importante incluir interacciones entre variables")
print("   - Validar outliers en precios muy altos/bajos")
```
